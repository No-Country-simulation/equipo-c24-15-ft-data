```python
#Instalamos las librerias necesarias

import pandas as pd
import numpy as np
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')  
```

```python
credit_risk_trainset = pd.read_csv("C:/equipo-c24-15-ft-data/datasets/cs-training.csv")

credit_risk_trainset.head()
```

```python
#Renombramos columna 

credit_risk_trainset.rename(columns = {credit_risk_trainset.columns[0]:'Id'}, inplace = True)
```

```python
#Eliminamos columnas innecesarias

credit_risk_trainset.drop(credit_risk_trainset.columns[[4, 8, 10]], axis=1, inplace=True)
credit_risk_trainset.head()
```

```python
# Reemplazamos valores >1 por NaN: Utilización de crédito no puede superar el 100%

credit_risk_trainset['RevolvingUtilizationOfUnsecuredLines'] = np.where(
    credit_risk_trainset['RevolvingUtilizationOfUnsecuredLines'] > 1,
    np.NaN,
    credit_risk_trainset['RevolvingUtilizationOfUnsecuredLines']
)
```

```python
# Rellenamos valores NaN con el último valor válido anterior (forward fill)

credit_risk_trainset['RevolvingUtilizationOfUnsecuredLines'].fillna(method='ffill', inplace=True)
```
